```python
import pandas as pd

# 1. Dosya isimlerini senin yÃ¼klediÄin isimlerle eÅleÅtiriyoruz
files = ['Aligned.csv', 'Flocking.csv', 'Grouped.csv']

# 2. DosyalarÄ± okuyalÄ±m
df_list = []
for f in files:
    try:
        # Bu dosyalarda genelde ilk satÄ±r boÅ olur, skiprows=[0] o yÃ¼zden var
        temp_df = pd.read_csv(f, skiprows=[0], encoding='latin1')
        df_list.append(temp_df)
        print(f"{f} baÅarÄ±yla okundu. Boyut: {temp_df.shape}")
    except Exception as e:
        print(f"{f} okunurken hata oluÅtu: {e}")

# 3. Hepsini tek bir tabloda birleÅtiriyoruz
df = pd.concat(df_list, ignore_index=True)

# 4. SÃ¼tun isimlerindeki boÅluklarÄ± temizleyelim
df.columns = df.columns.str.strip()

print("\n--- BÄ°RLEÅTÄ°RME TAMAMLANDI ---")
print(f"Toplam SatÄ±r SayÄ±sÄ±: {df.shape[0]}")
print(f"Toplam SÃ¼tun SayÄ±sÄ±: {df.shape[1]}")
```

```python
target_column = df.columns[-1]
# Verinin genel yapÄ±sÄ±nÄ± ve rastgele Ã¶rnekleri inceleyelim
print(f"Hedef SÃ¼tun AdÄ±: {target_column}")
print("-" * 30)
print("SÄ±nÄ±f DaÄÄ±lÄ±mÄ±:")
print(df[target_column].value_counts())
print("-" * 30)

# Rastgele 5 satÄ±rÄ± gÃ¶ster (TÃ¼m sÃ¼tunlarÄ± gÃ¶rmen iÃ§in .T yani devriÄini alalÄ±m, daha rahat okunur)
display(df.sample(5).T)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# --- 1. VERÄ° HAZIRLAMA ---
target_column = df.columns[-1]
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# SayÄ±sal temizlik (Dtype hatalarÄ±nÄ± Ã§Ã¶zer)
X = X.apply(pd.to_numeric, errors='coerce').fillna(0)

# --- 2. PCA (BOYUT AZALTMA) ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# --- 3. MODEL EÄÄ°TÄ°MÄ° ---
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# --- 4. SONUÃLAR VE GÃRSELLEÅTÄ°RME ---
y_pred = model.predict(X_test)

print("--- PROJE SONUÃLARI ---")
print(f"BaÅarÄ± OranÄ± (Accuracy): %{accuracy_score(y_test, y_pred)*100:.2f}")
print("\nSÄ±nÄ±flandÄ±rma Raporu:")
print(classification_report(y_test, y_pred))

# Hata Matrisi GrafiÄi
plt.figure(figsize=(6,4))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Greens')
plt.title('Hata Matrisi (Confusion Matrix)')
plt.xlabel('Tahmin Edilen')
plt.ylabel('GerÃ§ek DeÄer')
plt.show()

# PCA KÃ¼meleme GrafiÄi
plt.figure(figsize=(10, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='coolwarm', alpha=0.5)
plt.title('PCA ile SÃ¼rÃ¼ DavranÄ±ÅÄ± Analizi')
plt.colorbar(label='SÃ¼rÃ¼ Durumu')
plt.show()
```
